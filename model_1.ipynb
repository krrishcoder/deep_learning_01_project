```python
import numpy as np
from tensorflow.keras.datasets import mnist  # Using TensorFlow to load MNIST dataset

# Load the MNIST dataset
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Filter to include only labels 0 and 1
train_mask = (train_labels == 0) | (train_labels == 1)
test_mask = (test_labels == 0) | (test_labels == 1)

train_images = train_images[train_mask]
train_labels = train_labels[train_mask]
test_images = test_images[test_mask]
test_labels = test_labels[test_mask]

# Flatten images
train_images_flat = train_images.reshape(train_images.shape[0], -1)
test_images_flat = test_images.reshape(test_images.shape[0], -1)

# Normalize the data
train_images_normalized = train_images_flat / 255.0
test_images_normalized = test_images_flat / 255.0



# Ensure the data is in float32 for TensorFlow/Keras
train_images_normalized = train_images_normalized.astype('float32')
test_images_normalized = test_images_normalized.astype('float32')

# Print the shape of the normalized data (before reshaping)
print("Train images shape after normalization and flattening:", train_images_normalized.shape)
print("Test images shape after normalization and flattening:", test_images_normalized.shape)
```

```python
import numpy as np

def get_predictions(A3):
    new_array = np.zeros_like(A3)

    new_array[A3 >= 0.5] = 1
    

    return new_array

    
def get_accuracy(predictions, yyy):
    
    return (np.sum(predictions ==yyy)) / yyy.size


def ccv(arr):
    if arr.ndim == 1:
        return arr.reshape(-1, 1)
    return arr

def init_parms():
    # w1 is (3,m) or (m,3) , 3 bcoz first layer has 3 nodes and total features is 784

    w1 = np.random.randn(128, 784) * np.sqrt(2 / (128 + 784))
    b1 = np.zeros((128, 1))

    w2 = np.random.randn(64, 128) * np.sqrt(2 / (64 + 128))
    b2 = np.zeros((64, 1))

    w3 = np.random.randn(1, 64) * np.sqrt(2 / (1 + 64))
    b3 = np.zeros((1, 1))

    return w1,b1 , w2,b2, w3,b3

def sigmoid(Z):
    Z = Z.astype(np.float64)
    return 1/ (1 + np.exp(-Z))

def forward_propagation(w1,b1 , w2,b2, w3,b3 , X):



    Z1 = w1.dot(X)  # note that here X is only a single row of whole data


    Z1 = Z1 + b1

    A1 = sigmoid(Z1)



    Z2 = w2.dot(A1)
    Z2 = Z2 + b2

    A2 = sigmoid(Z2)
   
    Z3 = w3.dot(A2)
    Z3 = Z3 + b3

    A3 = sigmoid(Z3)


    Z1= ccv(Z1)
    A1 = ccv(A1)
    
    A2 = ccv(A2)
    Z2 =  ccv(Z2)
    
    Z3= ccv(Z3) 
    A3 = ccv(A3)


    return Z1,A1, Z2, A2 ,Z3,A3

def back_prop( Z1,A1, Z2 ,A2 , Z3,A3 ,w3, w2,Y , X):
    # here Y is (1,1)
    m = 12665

    dP3 = A3 - Y 
    dW3 = 1/m * dP3.dot(A2.T)
    dB3 = 1/m * np.sum(dP3, axis=1, keepdims=True)

    
    
    temp = A2*(1-A2)
    
    dP2  = w3.T.dot(dP3)
    dP23 = dP2 * temp
    

    dW2 = 1/m * (dP23.dot(A1.T))
    dB2 = 1/m * np.sum(dP23 , 1,keepdims=True)
  
    
    
    # use dP2
   
    dP14 = A1*(1-A1)
    tempp = dP2 * temp
    
    temp2 = w2.T.dot(tempp)
    temp2 = temp2 * dP14
    
    dW1 = 1/m * (temp2.dot(X.T))

    ## (3, 1) (1, 1000) (3, 1000) (1000, 784)
    
    dB1 = 1/m * np.sum(temp2,1, keepdims=True)

    dW1 = ccv(dW1)
    dB1 = ccv(dB1)
    dW2 = ccv(dW2)
    dB2 = ccv(dB2)
    dW3 = ccv(dW3)
    dB3 = ccv(dB3)


    return dW1, dB1 , dW2 , dB2 , dW3 , dB3


def update_parmas(w1,b1 , w2,b2, w3,b3, dW1, dB1 , dW2 , dB2 , dW3 , dB3, alpha):

    

    w1 = w1 - (alpha * dW1)
    b1 =  b1 - (alpha * dB1)

  

    w2 = w2 - (alpha * dW2)
    b2 = b2 - (alpha * dB2)

    w3 = w3 - (alpha * dW3)
    b3 =  b3 - (alpha * dB3)

    return w1,b1 , w2,b2, w3,b3


def gradient_des(X,Y , iterations ,alpha ):
    w1,b1 , w2,b2, w3,b3 = init_parms()

    for i in range(iterations):
         print("forward propagate")
         Z1,A1, Z2,A2 ,Z3,A3 = forward_propagation(w1,b1 , w2,b2, w3,b3 , X)
       
         dW1, dB1 , dW2 , dB2 , dW3 , dB3 = back_prop(Z1,A1, Z2 ,A2 , Z3,A3 ,w3,w2,Y , X)

         w1,b1 , w2,b2, w3, b3 = update_parmas(w1,b1 , w2,b2, w3,b3, dW1, dB1 , dW2 , dB2 , dW3 , dB3, alpha)

        
         print("iterations : ",i)
         print("accuracy : " , get_accuracy(get_predictions(A3) ,Y) )
         
      
        
    return w1,b1 , w2,b2, w3,b3
         

```

```python
import cv2
import numpy as np




# Initialize the webcam
cap = cv2.VideoCapture(0)

top_left_corner = (150, 150)
bottom_right_corner = (178, 178)
thickness = 2
color = (0, 255, 0)

flag = 1

frames = np.zeros((784, 3))

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    if not ret:
        break

    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)



    x, y, w, h = 150, 150, 28, 28  # Adjust these values as needed

    cropped_frame = gray[y:y+h, x:x+w]

    arr = np.array(cropped_frame)


    thresholded_image = np.where(arr > 150, 0, arr)

    flat_img = thresholded_image.reshape(-1)

    # my_images[:,0]  = flat_img
    # my_images[:,1]  = flat_img

    gray[y:y+h, x:x+w] = thresholded_image


    frames[:, flag%3] =  flat_img
  
    str = "prediction is : "

    if flag%3 == 0 :

        frames = frames /255.0
        frames = frames.astype('float32')

        print(frames.shape)
        Z11 ,A11, Z22,A22 ,Z33,A33 = forward_propagation(w1,b1 , w2,b2, w3,b3 , frames)
        #print("result is : ", get_predictions(A33))
        str += np.array2string(get_predictions(A33) , separator=', ')


    flag +=1

    cv2.putText(gray , str , (x-20,y-20) , cv2.FONT_HERSHEY_SIMPLEX, 1 , (0, 0, 0) , 2 )

    
    
    cv2.rectangle(gray, top_left_corner, bottom_right_corner, color, thickness)
    cv2.imshow('Digit Recognition', gray)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything is done, release the capture
cap.release()
cv2.destroyAllWindows()
```
