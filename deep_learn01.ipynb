```python
def get_predictions(A3):
    new_array = np.zeros_like(A3)

    new_array[A3 >= 0.5] = 1

    return new_array

    
def get_accuracy(predictions, Y):
    return (np.sum(predictions ==Y) / Y.size)


def ccv(arr):
    if arr.ndim == 1:
        return arr.reshape(-1, 1)
    return arr

def init_parms():
    # w1 is (3,m) or (m,3) , 3 bcoz first layer has 3 nodes and total features is 784

    w1 = np.random.randn(3, 784) - 0.5
    b1 = np.random.randn(3, 1) - 0.5

    w2 = np.random.randn(2, 3) - 0.5
    b2 = np.random.randn(2, 1) - 0.5

    w3 = np.random.randn(1, 2) - 0.5
    b3 = np.random.randn(1, 1) - 0.5

    return w1,b1 , w2,b2, w3,b3

def sigmoid(Z):
    Z = Z.astype(np.float128)
    return 1/ (1 + np.exp(-Z))

def forward_propagation(w1,b1 , w2,b2, w3,b3 , X):



    Z1 = w1.dot(X)  # note that here X is only a single row of whole data


    Z1 = Z1 + b1

    A1 = sigmoid(Z1)



    Z2 = w2.dot(A1)
    Z2 = Z2 + b2

    A2 = sigmoid(Z2)
   
    Z3 = w3.dot(A2)
    Z3 = Z3+ b3

    A3 = sigmoid(Z3)


    Z1= ccv(Z1)
    A1 = ccv(A1)
    Z2 =  ccv(Z2)
    Z3= ccv(Z3) 
    A3 = ccv(A3)


    return Z1,A1, Z2,A2 ,Z3,A3

def back_prop( Z1,A1, Z2 ,A2 , Z3,A3 ,w3,w2,Y , X):
    # here Y is (1,1)
    m = 1

    dP3 = A3 - Y 
    dW3 = 1/m * dP3.dot(A2.T)
    dB3 = 1/m * np.sum(dP3,1)

    temp = A2*(1-A2)

    dP2 = (w3.T)*temp

    dP23 = dP3.dot(A1.T)


    dW2 = 1/m * (dP2.dot(dP23))


    dB2 = 1/m * np.sum(dP2.dot(dP3), 1)

    

    dP1 = A3 - Y 
    dP12 = dP2.dot(dP3)
    dP13 =  w2.T.dot(dP12)
    dP14 = A1*(1-A1)
    dP15  = dP13*dP14

    dW1 = 1/m * (dP15.dot(X.T))
    dB1 = 1/m * np.sum(dP15,1)

    dW1 = ccv(dW1)
    dB1 = ccv(dB1)
    dW2 = ccv(dW2)
    dB2 = ccv(dB2)
    dW3 = ccv(dW3)
    dB3 = ccv(dB3)


    return dW1, dB1 , dW2 , dB2 , dW3 , dB3


def update_parmas(w1,b1 , w2,b2, w3,b3, dW1, dB1 , dW2 , dB2 , dW3 , dB3, alpha):

    

    w1 = w1 - alpha * dW1
    b1 =  b1 - alpha * dB1

  

    w2 = w2 - alpha * dW2

    b2 = b2 - (alpha * dB2)

    w3 = w3 - alpha * dW3
    b3 =  b3 - alpha * dB3

    return w1,b1 , w2,b2, w3,b3


def gradient_des(X,Y , iterations ,alpha ):
    w1,b1 , w2,b2, w3,b3 = init_parms()

    for i in range(iterations):
         print("forward propagate")
         Z1,A1, Z2,A2 ,Z3,A3 = forward_propagation(w1,b1 , w2,b2, w3,b3 , X)
       
         dW1, dB1 , dW2 , dB2 , dW3 , dB3 = back_prop(Z1,A1, Z2 ,A2 , Z3,A3 ,w3,w2,Y , X)

         w1,b1 , w2,b2, w3,b3 = update_parmas(w1,b1 , w2,b2, w3,b3, dW1, dB1 , dW2 , dB2 , dW3 , dB3, alpha)

        
         print("iterations : ",i)
         print("accuracy : ", get_accuracy(get_predictions(A3), Y))
        
    return w1,b1 , w2,b2, w3,b3
         

```
